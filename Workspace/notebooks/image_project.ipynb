## 2. Dataset Loading

```python
# Default to the smaller Garbage dataset for faster iteration.
# Food requires combining food1 and food2 into one folder before training.
GARBAGE_DATASET_DIR = r"C:\Users\evang\Downloads\garbage\garbage"
FOOD1_DATASET_DIR = r"C:\Users\evang\Downloads\food1\food1"
FOOD2_DATASET_DIR = r"C:\Users\evang\Downloads\food2\food2"

DATASET_DIR = GARBAGE_DATASET_DIR
IMAGE_SIZE  = (224, 224)
BATCH_SIZE  = 32
SEED = 419296

tf.random.set_seed(SEED)
np.random.seed(SEED)

train_ds, val_ds, test_ds, class_names = load_datasets(
    DATASET_DIR,
    image_size=IMAGE_SIZE,
    batch_size=BATCH_SIZE,
    seed=SEED,
    train_ratio=0.7,
    val_ratio=0.1,
    test_ratio=0.2,
    color_mode="rgb",
)

train_ds = prepare_dataset(train_ds, training=True, seed=SEED)
val_ds = prepare_dataset(val_ds)
test_ds = prepare_dataset(test_ds)

NUM_CLASSES = len(class_names)
print("Classes:", class_names)
print("Number of classes:", NUM_CLASSES)
```

## 3. Exploratory Data Analysis

```python
class_counts = count_images_per_class(DATASET_DIR)
display(pd.DataFrame.from_dict(class_counts, orient="index", columns=["count"]))

fig = plot_class_distribution(class_counts)
plt.show()

images, labels = next(iter(train_ds.take(1)))
print("Batch image shape:", images.shape)
print("Pixel range:", float(tf.reduce_min(images)), "to", float(tf.reduce_max(images)))

plt.figure(figsize=(10, 10))
for i in range(min(9, images.shape[0])):
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(images[i].numpy().astype("uint8"))
    plt.title(class_names[int(labels[i])])
    plt.axis("off")
plt.tight_layout()
plt.show()

class_weight = compute_class_weights(class_counts)
print("Class weights:", class_weight)
```

## 4. Dataset Splitting

```python
print("Train batches:", train_ds.cardinality().numpy())
print("Validation batches:", val_ds.cardinality().numpy())
print("Test batches:", test_ds.cardinality().numpy())
print("Split used: 70% train, 10% validation, 20% test")
```
